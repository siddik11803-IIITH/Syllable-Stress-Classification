```python
import scipy.io as sio
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import pickle as pkl
```

```python
GER_test = np.array(sio.loadmat('./../GER_test.mat')['GER_test'])[2:].transpose()
GER_test_y = np.array(sio.loadmat('./../GER_test.mat')['GER_test'])[0].transpose()
GER_train =np.array(sio.loadmat('./../GER_train.mat')['GER_train'])[2:].transpose()
GER_train_y =np.array(sio.loadmat('./../GER_train.mat')['GER_train'])[0].transpose()
ITA_train =np.array(sio.loadmat('./../ITA_train.mat')['ITA_train'])[2:].transpose()
ITA_train_y =np.array(sio.loadmat('./../ITA_train.mat')['ITA_train'])[0].transpose()
ITA_test = np.array(sio.loadmat('./../ITA_test.mat')['ITA_test'])[2:].transpose()
ITA_test_y = np.array(sio.loadmat('./../ITA_test.mat')['ITA_test'])[0].transpose()
```
